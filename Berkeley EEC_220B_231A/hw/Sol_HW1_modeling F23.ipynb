# <font color=blue> Question 1.  Equilibrium Point and Linearization </font>

**Part (a)**

At equilibrium $\dot{\theta}_1=\dot{\theta}_2=\ddot{\theta}_1=\ddot{\theta}_2=0$. We set $\bar{\theta}_2=0$ and get
    
\begin{align}
mgl\sin\theta_1-a^2k\cos\theta_1\sin\theta_1+T(t) &= 0\\
a^2k\sin\theta_1+\alpha                           &= 0
\end{align}
    
After some rearrangement we get
    
\begin{align}
\bar{\theta}_1 &= \arcsin{\left(-\frac{\alpha}{a^2k}\right)}\\
\bar{T}        &= mgl\frac{\alpha}{a^2k}-\alpha\cos\arcsin{\left(-\frac{\alpha}{a^2k}\right)}
\end{align}
    
Using $\cos\arcsin{(x)}=\sqrt{1-x^2}$, the equation for $\bar{T}$ can be simplified to
    
\begin{align}
\bar{T}        &= \frac{\alpha}{a^2k}\left(mgl-\sqrt{a^4k^2-\alpha^2}\right)
\end{align}

**Part (b)**

The new equilibrium point is $\bar{\theta}_1=\bar{\theta}_2=\bar{T}=0$. We set the state and input variables as follows
    
\begin{align}
\mathbf{x} &= \begin{bmatrix}
           x_1\\x_2\\x_3\\x_4
          \end{bmatrix} =
  \begin{bmatrix}
   \theta_1-\bar{\theta}_1=\theta_1\\
   \dot{x}_1\\
   \theta_2-\bar{\theta}_2=\theta_2\\
   \dot{x}_2
  \end{bmatrix}\:,\qquad
  u = T(t)-\bar{T}=T(t)
\end{align}
    
The system of four $1$st order ODE $\dot{x}=f(x,u)$ can be written as
    
\begin{align}
f(x,u) &= \begin{bmatrix}
       x_2\\
   \frac{1}{ml^2}\left(mgl\sin x_1 + a^2\cos x_1\left(k(\sin x_3-\sin x_1)+d(x_4-x_2)\right)+u\right)\\
   x_4\\
   \frac{1}{ml^2}\left(mgl\sin x_3 - a^2\cos x_3\left(k(\sin x_3-\sin x_1)+d(x_4-x_2)\right)+\alpha+\beta x_4^2\right)
      \end{bmatrix}
\end{align}
    
Taking the hint into account and linearizing with
    
\begin{align}
A &= \left.\frac{\partial f(x,u)}{\partial x}\right|_{x_{ss},u_{ss}}\:, \qquad
B = \left.\frac{\partial f(x,u)}{\partial u}\right|_{x_{ss},u_{ss}}
\end{align}
    
leads to a state space description of the form $\dot{x}=Ax+Bu$ given by
    
\begin{align}
\dot{x}(t) &= \begin{bmatrix}
               0&1&0&0\\
       \frac{g}{l}-\frac{a^2k}{ml^2} & -\frac{a^2d}{ml^2} & \frac{a^2k}{ml^2} & \frac{a^2d}{ml^2}\\
       0&0&0&1\\
       \frac{a^2k}{ml^2} & \frac{a^2d}{ml^2} & \frac{g}{l}-\frac{a^2k}{ml^2} & -\frac{a^2d}{ml^2}
              \end{bmatrix} x(t)+
      \begin{bmatrix}
       0\\\frac{1}{ml^2}\\0\\0
      \end{bmatrix} u(t)\\
y(t)  &= \begin{bmatrix}
               1&0&0&0
              \end{bmatrix} x(t)+ 0 \cdot u(t)
\end{align}

```python
def EquilPoint(theta2, alpha):
    a = 3
    k= 1.5
    m = 2
    g = 9.81
    l = 6
    d = 1
    beta = 0.1
    temp = (-m*g*l*sin(theta2)-alpha+a**2*k*cos(theta2)*sin(theta2)) / (a**2*k*cos(theta2) )
    theta1 = arcsin(temp)
    T = -m*g*l*sin(theta1)-a**2*k*cos(theta1)*(sin(theta2)-sin(theta1))
    return  theta1, T

# For example for the fixed values of theta2 = 0.1 and alpha = 0.5
theta1_bar, T_bar = EquilPoint(theta2=0.1, alpha=0.5)
print(theta1_bar)
print(T_bar)
```

```python
import sympy as sym

def LinearizeModel(theta2, alpha):

    a = 3
    k = 1.5
    m = 2
    g = 9.81
    l = 6
    d = 1
    beta = 0.1
    theta1dot = 0
    theta2dot = 0

    theta1, T = EquilPoint(theta2, alpha)  # calling the above EquilPoint function

    x1, x2, x3, x4, u1, u2 = sym.symbols('x1 x2 x3 x4 u1 u2')
    f = sym.Matrix([x2, (m*g*l*sym.sin(x1)+a**2*sym.cos(x1)*(k*(sym.sin(x3)-sym.sin(x1))+d*(x4-x2))+u1)/(m*l**2),
                x4, (m*g*l*sym.sin(x3)-a**2*sym.cos(x3)*(k*(sym.sin(x3)-sym.sin(x1))+d*(x4-x2))+u2+beta*x4**2)/(m*l**2)])
    Asym = f.jacobian([x1, x2, x3, x4])
    print(Asym)
    Asub = Asym.subs([(x1, theta1), (x2, theta1dot), (x3, theta2), (x4, theta2dot), (u1, T), (u2, alpha)])

    Bsym = f.jacobian([u1, u2])
    Bsub = Bsym.subs([(x1, theta1), (x2, theta1dot), (x3, theta2), (x4, theta2dot), (u1, T), (u2, alpha)])

    C = np.array([1, 0, 0, 0])
    D = np.array([0, 0])

    return Asub, Bsub, C, D, theta1, T

# Please submit the results of the following unit test
theta2_bar = 0.1
alpha_bar = 0.5
A, B, C, D, theta1_bar, T_bar = LinearizeModel(theta2_bar, alpha_bar)
print('A = ', A)
print('B = ', B)
print('C = ', C)
print('D = ', D)
print('theta1_bar = ', theta1_bar)
print('T_bar = ', T_bar)
```

#  <font color=blue> Question 2. Euler Discretization of a Building Heat Transfer Model </font>
**Part (a)**

Starting with our original equation:

\begin{align}
m_z c_z \dot{T} = q + c_pu_1(u_2 - T) \nonumber
\end{align}

Use the Euler discretization approximation:

\begin{align}
\dot{T} = \frac{T(k+1)-T(k)}{T_s} \nonumber
\end{align}

Substitute in:
$$
m_z c_z \frac{T(k+1)-T(k)}{T_s} = q(k)+ c_pu_1(k)(u_2(k) - T(k)) \nonumber
$$

Reshuffle variables:
$$
T(k+1) = \left[ 1 - \frac{c_p T_s}{m_z c_z} u_1(k) \right]T(k) + \frac{T_s}{m_zc_z}\left[q(k) + c_pu_1(k)u_2(k)\right] \nonumber
$$


```python
def bldgHTM(T, u1, u2, q, mz, cz, cp):
    Tdot = (q+cp*u1*(u2-T))/(mz*cz)
    return Tdot


Tdot = bldgHTM(T=60, u1=0.001, u2 = 10, q=1000,  mz=100, cz=20, cp=1000)
print(Tdot)
```

**Part (b)**

```python
def eulerDiscretization(T,q,u1,u2,Ts):
    mz = 100
    cz = 20
    cp = 1000
    T_KplusOne = (1-cp*Ts/(mz*cz)*u1)*T+Ts/(mz*cz)*(q+cp*u1*u2)
    return T_KplusOne

T_KplusOne = eulerDiscretization(T=60, q=1000, u1=0.001, u2 = 100, Ts = 0.1)
print(T_KplusOne)
```

**Part (c)**

**System stability:** Systems that are marginally stable likely require higher sampling times. Even if a system is stable to begin with, a too large time step an render the system unstable.

**Available computation:** Smaller timesteps require more computation, since the system is being simulated at more instances.

**Input rate:** If the input changes slowly, we can tolerate longer timesteps. If the system is highly dynamic, and the input rate is high, then large time steps can result in instability. As we saw with ZOH discretization for linear systems, ZOH is an exact discretization of the system dynamics as long as the input is constant over the sampling time.

```python
import matplotlib.pyplot as plt

def carModel(beta, a, x, y, v, psi):

    l_r   = 1.738
    dt = 0.1

    x_dot   = v*cos(psi+beta)
    y_dot   = v*sin(psi+beta)
    psi_dot = (v/l_r)*sin(beta)
    v_dot   = a

    x_out   = x + x_dot*dt
    y_out   = y + y_dot*dt
    psi_out = psi + psi_dot*dt
    v_out   = v + v_dot*dt

    return x_out, y_out, psi_out, v_out

# Calling the function example
x, y, psi, v = carModel(0.1,2,5,2,0.1,10)
print(x)
print(y)
print(v)
print(psi)
```

# <font color=blue> Question 5. Linearization and Discretization </font>


We want to go from some nonlinear continous-time system:

\begin{align}
\dot x^c(t) = g^c(x(t),u(t))
\end{align}

to a lineared, discrete time model of the system:

\begin{align}
x(k+1) = Ax(k) + Bu(k)
\end{align}

To do so we have two steps--linearization and discretization. We are interested in determining if the order in which these steps are preformed matters. *Note:* In this exercise we are considering Jacobian linearization and forward Euler discretization, the answer to this question will depend on the method of linearization and discretization.

**Linearization First Approach:**

Begin with the nonlinear continous-time system dynamics:

\begin{align}
\dot x^c(t) = g^c(x(t),u(t))
\end{align}

Linearize about point $(x_s, u_s)$

\begin{align}
\dot x^c(t) \approx g^c(x_s, u_s) + \left.\frac{\partial g^c}{\partial x}\right|_{x_{s},u_{s}} (x(t)-x_s) + \left.\frac{\partial g^c}{\partial u}\right|_{x_{s},u_{s}} (u(t)-u_s)
\end{align}

\begin{align}
\Delta \dot x^c(t) &= \left.\frac{\partial g^c}{\partial x}\right|_{x_{s},u_{s}} \Delta x(t) + \left.\frac{\partial g^c}{\partial u}\right|_{x_{s},u_{s}} \Delta u(t) \\
&= A^c \Delta x(t) + B^c \Delta u(t)
\end{align}

where

\begin{align}
A^c = \frac {\delta g^c}{\delta x} \bigg \rvert _{x_s, u_s}, \qquad
B^c = \frac {\delta g^c}{\delta u} \bigg \rvert _{x_s, u_s}
\end{align}


Now discretizing with forward Euler:
\begin{align}
\Delta x(k+1) &= \Delta x(k) (I + T_s A^c) + \Delta u(k) T_s B^c \\
&= A^d \Delta x(k) + B^d \Delta u(k)
\end{align}

where

\begin{align}
A^d = I + T_s A^c, \qquad
B^d = T_s B^c
\end{align}

**Discretization First Approach:**
Begin with the nonlinear continous-time system dynamics:

\begin{align}
\dot x^c(t) = g^c(x(t),u(t))
\end{align}

Discretize:

\begin{align}
x(k+1) = x(k) + T_s g^c(x^c(t), u^c(t))
\end{align}

Now linearize about $(x_s, u_s)$. *Note:* we are using the same equilibrium point to linearize about in both approaches.

\begin{align}
x(k+1) \approx x_s + T_s g^c(x_s, u_s) + \left.\frac {\partial}{\partial x} (x + T_s g^c(x,u)) \right|_{x_s, u_s} (x(k)-x_s) + \left.\frac {\partial}{\partial u} (x + T_s g^c(x,u)) \right|_{x_s, u_s} (u(k)-u_s)
\end{align}

\begin{align}
\Delta x(k+1) &= x(k+1) - x_s \\
&\approx (I+T_s \left.\frac {\partial g^c}{\partial x} \right |_{x_s, u_s}) (x(k)-x_s) + T_s \frac {\partial g^c} {\partial u} (u(k)-u_s) \\
&= \Delta x(k)(I+T_s A^c) + \Delta u(k) T_s B^c \\
&= A^d \Delta x(k) + B^d \Delta u(k)
\end{align}

where

\begin{align}
A^d = I + T_s A^c, \qquad
B^d = T_s B^c
\end{align}

For Jacobian linearization and forward Euler discretization, the order of linearization and discretization does not matter.
